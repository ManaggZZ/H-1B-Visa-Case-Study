## Data Cleaning

```python
df.columns = df.columns.str.strip().str.upper().str.replace(' ', '_')
```

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
print(missing_values[missing_values > 0])
```

```python
# Drop rows with missing case status, job title, or wage
df.dropna(subset=['PETITIONER_ZIP_CODE', 'PETITIONER_STATE', 'PETITIONER_CITY', 'INDUSTRY_(NAICS)_CODE','TAX_ID', 'EMPLOYER_(PETITIONER)_NAME'], inplace=True, ignore_index=True)
```

```python
df.drop(columns = ['LINE_BY_LINE', 'FISCAL_YEAR'], inplace=True)
```

```python
df.reindex()
```

```python
df['CONTINUATION_APPROVAL'] = pd.to_numeric(df['CONTINUATION_APPROVAL'], errors='coerce')
df['CHANGE_WITH_SAME_EMPLOYER_APPROVAL'] = pd.to_numeric(df['CHANGE_WITH_SAME_EMPLOYER_APPROVAL'], errors='coerce')
df['AMENDED_APPROVAL'] = pd.to_numeric(df['AMENDED_APPROVAL'], errors='coerce')
df = df.dropna(subset=['CONTINUATION_APPROVAL', 'CHANGE_WITH_SAME_EMPLOYER_APPROVAL', 'AMENDED_APPROVAL'])
```

```python
df.info()
```

```python
# Convert count columns to numeric
approval_cols = [col for col in df.columns if col.endswith('_APPROVAL')]
denial_cols = [col for col in df.columns if col.endswith('_DENIAL')]
df[approval_cols + denial_cols] = df[approval_cols + denial_cols].apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
```

```python
# Add total columns
df['TOTAL_APPROVED'] = df[approval_cols].sum(axis=1)
df['TOTAL_DENIED'] = df[denial_cols].sum(axis=1)
df['TOTAL_FILED'] = df['TOTAL_APPROVED'] + df['TOTAL_DENIED']
df['APPROVAL_RATE'] = (df['TOTAL_APPROVED'] / df['TOTAL_FILED']).round(4)
```

```python
df
```

### EDA

```python
# Top Employers by Petition Volume
top_employers = df.groupby('EMPLOYER_(PETITIONER)_NAME')['TOTAL_FILED'].sum().sort_values(ascending=False).head(10)
print(top_employers)
```

```python
# Employers with Highest Approval Rates (with 50 filings)
qualified = df[df['TOTAL_FILED'] >= 50]
top_approval_rates = qualified.groupby('EMPLOYER_(PETITIONER)_NAME')['APPROVAL_RATE'].mean().sort_values(ascending=False).head(10)
print(top_approval_rates)
```

```python
# # Total New Employment Petitions by Year
# df.groupby('FISCAL_YEAR')[['NEW_EMPLOYMENT_APPROVAL', 'NEW_EMPLOYMENT_DENIAL']].sum()
```

```python
# Top 5 States by Total Petitions
top_states = df.groupby('PETITIONER_STATE')['TOTAL_FILED'].sum().sort_values(ascending=False).head(5)
print(top_states)
```

```python
# Most Common Petition Types Filed
petition_type_totals = df[approval_cols + denial_cols].sum().sort_values(ascending=False)
print(petition_type_totals)
```

```python
import matplotlib.pyplot as plt
petition_type_totals.plot(kind='barh', figsize=(10, 6), title='Total Petitions by Type (All Years)')
plt.xlabel('Number of Petitions')
plt.show()
```

```python
# Petition Type Approval Rates
petition_types = []

for app_col in approval_cols:
    den_col = app_col.replace('_APPROVAL', '_DENIAL')
    total = df[app_col].sum() + df[den_col].sum()
    rate = df[app_col].sum() / total if total > 0 else 0
    petition_types.append((app_col.replace('_APPROVAL', ''), rate, total))

petition_types_df = pd.DataFrame(petition_types, columns=['Petition_Type', 'Approval_Rate', 'Total_Filed'])
petition_types_df.sort_values(by='Approval_Rate', ascending=False, inplace=True, ignore_index=True)
print(petition_types_df)
```

```python
# Employers With Most Denials
top_denied = df.groupby('EMPLOYER_(PETITIONER)_NAME')['TOTAL_DENIED'].sum().sort_values(ascending=False).head(10)
print(top_denied)
```

```python
# Approval Rate by Petition Type
petition_types_df.plot(x='Petition_Type', y='Approval_Rate', kind='bar', figsize=(10, 5), color='green')
plt.title('Approval Rate by Petition Type')
plt.ylabel('Approval Rate')
plt.ylim(0, 1.1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Employers by Total Petitions Filed
top_employers = df.groupby('EMPLOYER_(PETITIONER)_NAME')['TOTAL_FILED'].sum().sort_values(ascending=False).head(10)

top_employers.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title('Top 10 Employers by Total Petitions Filed')
plt.xlabel('Total Petitions')
plt.gca().invert_yaxis()  # Highest at top
plt.tight_layout()
plt.show()
```

```python
# Petition Volume by State
import seaborn as sns
state_volume = df.groupby('PETITIONER_STATE')['TOTAL_FILED'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=state_volume.values, y=state_volume.index, palette='viridis', hue=0.6, legend=False)
plt.title('Top 10 States by Petition Volume')
plt.xlabel('Total Petitions Filed')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```
